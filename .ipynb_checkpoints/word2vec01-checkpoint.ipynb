```python
from gensim.models import Word2Vec, KeyedVectors
import pandas as pd
import nltk
import os
os.getcwd()
```

```python
df = pd.read_csv('NYTCommentsMarch2018.csv')
```

```python
df.head(10)
```

```python
articles = df["commentBody"].values
```

```python
#articles
```

```python
#only once
#nltk.download('punkt')
```

```python
newsVec = [nltk.word_tokenize(article) for article in articles]
```

```python
newsVec
```

```python
model = Word2Vec(newsVec,min_count=1,size=32)
```

```python
model.most_similar('man')
```

```python
model.wv['man']
```

```python
vec2 = model["Germany"] - model["Berlin"] + model["Paris"]
model.most_similar([vec2])
```
